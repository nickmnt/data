# Loading the dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.2f}'.format
```

```python
df = pd.read_csv('movies_complete.csv', parse_dates=["release_date"])
```

# Overall analysis

```python
df.sort_values('revenue_musd', ascending=False).head(5)['title']
```

Highest budget

```python
df.sort_values('budget_musd', ascending=False).head(5)['title']
```

Highest Profit (Revenue - Budget)

```python
df['profit'] = df['revenue_musd'] - df['budget_musd']
```

```python
df.sort_values('profit', ascending=False).head(5)['title']
```

Lowest Profit (Revenue - Budget)

```python
df.sort_values('profit').head(5)['title']
```

Highest ROI (Revenue / Budget, only movies with budget >= 10)

```python
df1 = df[df['budget_musd'] >= 10]
df['ROI'] = df1['revenue_musd'] / df1['budget_musd']
```

```python
df.sort_values('ROI', ascending=False).head(5)['title']
```

Highest ROI (Revenue / Budget, only movies with budget >= 10)

```python
df.sort_values('ROI').head(5)['title']
```

Highest number of votes

```python
df.sort_values('vote_count', ascending=False).head(5)['title']
```

Highest rating (Only movies with more than 10 ratings)

```python
df2 = df[df['vote_count'] >= 10]
df2.sort_values('vote_average', ascending=False).iloc[0]['title']
```

Lowest rating (Only movies with more than 10 ratings)

```python
df2.sort_values('vote_average').head(5)['title']
```

Highest popularity

```python
df2.sort_values('popularity', ascending=False).head(5)['title']
```

# Finding specific movies

Science fiction action movies with Bruce Willis (sorted from high to low rating)

```python
df['genres']
```

```python
df['cast']
```

```python
df3 = df[df['genres'].notna() & df['cast'].notna()]
df3[df3['genres'].str.lower().str.contains('action') & df3['cast'].str.lower().str.contains('bruce willis')]
```

Movies with Uma Thurman and directed by Quentin Tarantino (sorted from short to long runtime)

```python
df4 = df[df['director'].notna() & df['cast'].notna()]
df4[df4['director'].str.lower().str.match('quentin tarantino') & df4['cast'].str.lower().str.contains('uma thurman')].sort_values('runtime')
```

Most successful Pixar Studio movies between 2010 and 2015 (sorted from high to low Revenue)

```python
df['release_date']
```

```python
df['release_date'].apply(lambda x: str(x).split("-")[0])
```

```python
df['release_date_year'] = df['release_date'].apply(lambda x: x.year)
df5 = df[df['production_companies'].notna() & df['release_date_year'].notna()]
df5[(df5['release_date_year'] <= 2015) & (df5['release_date_year'] >= 2010) & (df5['production_companies'].str.lower().str.match('pixar studio|'))].sort_values('revenue_musd',ascending=False)
```

Action or thriller movie with original language English and minimum rating 7.5 (most recent movies first)

```python
df6 = df[df['original_language'].notna() & df['genres'].notna() & df['vote_average'].notna()]
df6[df6['genres'].str.lower().str.contains('action|thriller') & df6['original_language'].str.match('en') & (df6['vote_average'] >= 7.5)].sort_values('release_date', ascending=False)
```

# Are franchises (belongs to collection) more successful?

Mean
* revenue
* ROI
* budget
* popularity
* rating

```python
df['is_franchise'] = ~df['belongs_to_collection'].isnull()
```

```python
df.groupby('is_franchise').mean(numeric_only=True)[['revenue_musd','budget_musd','ROI', 'popularity', 'vote_average']]
```

# Most successful franchises

* number of movies
* total and mean budget
* total and mean revenue
* mean rating

```python
df.groupby('belongs_to_collection').count()['id'].sort_values(ascending=False)
```

```python
df.groupby('belongs_to_collection').sum(numeric_only=True)['budget_musd'].sort_values(ascending=False)
```

```python
df.groupby('belongs_to_collection').mean(numeric_only=True)['budget_musd'].sort_values(ascending=False)
```

```python
df.groupby('belongs_to_collection').sum(numeric_only=True)['revenue_musd'].sort_values(ascending=False)
```

```python
df.groupby('belongs_to_collection').mean(numeric_only=True)['revenue_musd'].sort_values(ascending=False)
```

```python
df.groupby('belongs_to_collection').mean(numeric_only=True)['vote_average'].sort_values(ascending=False)
```
