## Addition of the ignore_index parameter

```python
import pandas as pd
```

```python
players = pd.read_csv("players.csv")
```

```python
players
```

```python
players.sort_values(by="Height", ascending=False)
```

```python
players.sort_values(by="Height", ascending=False).reset_index(drop=True) # Old way
```

```python
players.sort_values(by="Height", ascending=False, ignore_index=True) # New way
```

```python
players = pd.read_csv("players3.csv")
```

```python
players
```

```python
players.sort_values(by="Height", ascending=False)
```

```python
players.drop_duplicates()
```

```python
players.drop_duplicates(ignore_index=True)
```

## Removal of prior version deprecations

> ix is deprecated  

> players.loc[["Lionel Messi", "Donald Duck"]] worked in the past but doesn't now  
> reindex is used now, reindex only works with a unique index  

> ptp is deprecated, use x.max() - x.min() instead, you can also use np.ptp
