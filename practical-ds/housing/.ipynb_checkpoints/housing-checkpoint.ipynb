__Add__ the additional column __"income_cat"__ with the following income categories:
- lowest 25% -> "Low"
- 25th to 50th percentile -> "Below_Average"
- 50th to 75th percentile -> "Above_Average"
- 75th to 95th percentile -> "High"
- Above 95th percentile -> "Very High"

```python
pd.qcut(df.median_income, q = [0, 0.25, 0.5, 0.75, 0.95, 1], labels=['Low', 'Below_Average', 'Above_Average', 'High', 'Very High'])
```

```python
df['income_cat'] = pd.qcut(df.median_income, q = [0, 0.25, 0.5, 0.75, 0.95, 1], labels=['Low', 'Below_Average', 'Above_Average', 'High', 'Very High'])
```

```python
df.income_cat.value_counts(normalize=True)
```

## Machine Learning - Predicting House Values (optional)

Based on the Explanatory Data Analysis, <br>
- select an appropriate ML Model (e.g. Linear Regression or RandomForest Regression)
- split into Train and Test Set
- prepare/engineer Features
- Fit the Model (Train Set)
- Test the Model (Test Set)

```python
y = df['median_house_value'].copy()
```

```python
X = df.drop('median_house_value', axis=1)
```

```python
X.info()
```

```python
X.select_dtypes("float")
```

```python
import scipy.stats as stats
```

```python
feat1 = X.select_dtypes("float").apply(lambda x: stats.zscore(x))
feat1
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
feat1.agg(["mean", "std"])
```

```python
X['ocean_proximity']
```

```python
X['ocean_proximity'].value_counts()
```

```python
dummies = pd.get_dummies(X['ocean_proximity'])
dummies
```

```python
X = pd.concat([feat1, dummies, df['income_cat']], axis = 1)
X
```

```python
test_size = 0.2
```

```python
X_test = X.sample(frac = test_size, random_state= 123)
```

```python
X_test['income_cat'].value_counts(normalize = True)
```

```python
X['income_cat'].value_counts(normalize = True)
```

```python
X_test.index
```

```python
X_train = X.loc[~X.index.isin(X_test.index)].copy()
```

```python
X_train
```

```python
X_train['income_cat'].value_counts(normalize = True)
```

```python
X_train = X_train.sample(frac = 1, random_state= 123)
X_train
```

```python
X_train.drop(columns = ["income_cat"], inplace = True)
```

```python
X_test.drop(columns = ["income_cat"], inplace = True)
```

```python
y_train = y.loc[X_train.index]
y_test = y.loc[X_test.index]
```

```python
y_train
```

## Training the ML Model (Random Forest Regressor)

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest_reg = RandomForestRegressor(random_state = 42, n_estimators = 500,
                                   max_features= "sqrt", max_depth= 75, min_samples_split=2)
```

```python
forest_reg.fit(X_train, y_train)
```

```python
forest_reg.score(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
pred = forest_reg.predict(X_train)
pred
```

```python
import numpy as np
```

```python
forest_mse = mean_squared_error(y_train, pred)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

## Evaluating the Model on the Test Set

```python
forest_reg
```

```python
forest_reg.score(X_test, y_test)
```

```python
pred = forest_reg.predict(X_test)
pred
```

```python
forest_mse = mean_squared_error(y_test, pred)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

```python
comp = pd.DataFrame(data = {"True_V": y_test, "Pred":pred})
comp
```

```python
ae = (comp['True_V'] - comp['Pred']).abs()
ae
```

```python
mae = ae.mean()
mae
```
