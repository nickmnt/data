__Create__ the function __most_pop(year, gender, n)__ that returns the __n__ most popular __M/F__ Names in the __year XYZ__.

```python
def most_pop(year, gender, n):
    return df[(df['Year'] == year) & (df['Gender'] == gender)].nlargest(n, 'Count')

most_pop(2017, 'M', 10)
```

## Advanced Data Aggregation

```python
agg = df.groupby(['Name', 'Gender']).agg(Total=('Count', 'sum'), No_Year=('Year', 'count'), First_Year=('Year', lambda x: x.iloc[0]), Last_Year=('Year', lambda x: x.iloc[-1]), Max_Count=('Count', lambda x: max(x)))
agg
```

```python
def best_year(group):
    return group.nlargest(1, "Count").Year
```

## General Trends over time (1880 - 2018)

__Count__ the total number of __registered Babies per Year__ and visualize.

```python
baby_per_y = df.groupby('Year').agg(Total=('Count', 'sum'))
```

```python
import matplotlib.pyplot as plt

ax = baby_per_y.plot(kind='bar', figsize=(13, 10))

for i, t in enumerate(ax.get_xticklabels()):
    if (i % 5) != 0:
        t.set_visible(False)

plt.show()
```

__Count__ the total number of __Unique Names per Year__ and visualize.

```python
name_per_y = df.groupby('Year').agg(Total=('Name', 'count'))
```

```python
ax = name_per_y.plot(kind='bar', figsize=(13, 10))

for i, t in enumerate(ax.get_xticklabels()):
    if (i % 5) != 0:
        t.set_visible(False)

plt.show()
```

## Creating the Features "Popularity" and "Rank"

__Add__ the column __"Popularity"__ to df: <br>  
- Popularity in __babies per million__ (example: assume 2,000,000 registered Girl Names in 1950, thereof 100,000 Maries -> Popularity Score in 1950 for (Mary, F) is 100,000 / 2,000,000 * 1,000,000 = 50,000 per million.)   

__Add__ the column __"Rank"__ to df: <br>  
- Rank (example: assume Mary is the most popular Girl Name in 1950 -> Rank 1 for (Mary, F) in 1950.

```python
df['Popularity'] = df['Count'] / df.groupby(["Year", "Gender"]).Count.transform("sum") * 1e6
df['Popularity'] = df['Popularity'].astype(int)
```

```python
df
```

```python
df['Year']
```

```python
df
```

```python
df['Rank'] = df.groupby(['Year', 'Gender'])['Count'].rank(ascending=False)
```

```python
df
```

## Visualizing Name Trends over Time

 __Visualize__ Popularity and Rank over time (see example below).

```python
df1 = df[df['Name'] == 'Josh']
ax = df1.plot(x='Year', y='Popularity', kind='bar', figsize=(13, 10), color='Blue')
ax2 = ax.twinx()
df1.plot(x='Year', y='Rank', kind='line', ax=ax2, secondary_y=True, color='Red')

for i, t in enumerate(ax.get_xticklabels()):
    if (i % 10) != 0:
        t.set_visible(False)

plt.show()
```
